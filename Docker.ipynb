####Dockerfile :

```python
## Creating dockerfile : Dockerfile (no extension)
```

```python
## Keywords used to define Dockerfile instruction :

>> FROM : Specify base image (OS/Python Version)

>> COPY : Copy files from local to image

>> RUN : Execute command during buiding image (pip install)

>> CMD : Default commaand to run when container is started (python main.py/uvicorn main:app)

>> WORKDIR : Specify working directory
```

```python
## Set Base Image :
FROM python:3.10-slim

>> Set base image for container
```

```python
# !pwd
```

```python
## Set Working Directory :
WORKDIR /app

>> Create a folder in docker container
>> Set working directory to that folder
```

```python
## COPY source target
COPY requirements.txt .

>> Copy files from local to Image (Copy requirement.txt from local to image (app))
```

```python
## Install Dependencies :
RUN pip install -r requirements.txt

>> Install all dependencies from requirement.txt
```

```python
## Copy Code :
COPY . .

>> Copy all files from local to image
```

```python
## Run application :
# uvicorn main:app --host 0.0.0.0 --port 8080
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8080"]
```
